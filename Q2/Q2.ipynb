#### Reading data into Python and plotting
Download and install Breakthrough Listen I/O Methods for Python (blimpy) package.
https://blimpy.readthedocs.io/en/latest/contents.html
This tutorial gives necessary information about how to read data into Python using
blimpy. Following this tutorial, download Voyager 1 data and make a waterfall* plot with
frequencies between 8419.2740 to 8419.2750 MHz. Send us this plot.
* (if you do not understand what does “waterfall” mean, do not worry)

```python
import matplotlib.pyplot as plt
from blimpy import Waterfall
```

```python
!wget http://blpd0.ssl.berkeley.edu/Voyager_data/Voyager1.single_coarse.fine_res.h5
```

```python
file_path = 'Voyager1.single_coarse.fine_res.h5'
obs = Waterfall(file_path)
obs.info()
```

```python
%matplotlib inline
obs.plot_waterfall(f_start=8419.2740, f_stop=8419.2750)
plt.savefig("waterfall.png",dpi=300)
```
